Создайте класс по работе с тригонометрическими функциями. В классе
должны быть реализованы функции вычисления:
− косинуса;
− синуса;
− тангенса;
− арксинуса;
− арккосинуса;
− арктангенса;
− перевода из градусов в радианы.

```python
import numpy as np

class Trigonometry:

    def __init__(self):
        pass

    def cos(self, a):
        return np.cos(a)

    def sin(self, a):
        return np.sin(a)

    def tg(self, a):
        return np.tan(a)

    def arcsin(self, a):
        return np.arcsin(a)

    def arccos(self, a):
        return np.arccos(a)

    def arctg(self, a):
        return np.arctan(a)

    def toRadians(self, a):
        return np.radians(a)
```

```python
trig = Trigonometry()
results = []

results.append(trig.cos(np.pi/2))
results.append(trig.sin(np.pi/2))
results.append(trig.tg(np.pi/3))
results.append(trig.arcsin(1))
results.append(trig.arccos(0))
results.append(trig.arctg(0))
results.append(trig.toRadians(180))
results
```

Реализуйте в классе функцию для вставки нового элемента в дерево по
следующим правилам:
• Левое поддерево узла содержит только узлы со значениями меньше,
чем значение в узле.
• Правое поддерево узла содержит только узлы со значениями меньше,
чем значение в узле.
• Каждое из левого и правого поддеревьев также должно быть
бинарным деревом поиска.
• Не должно быть повторяющихся узлов.
Метод вставки сравнивает значение узла с родительским узлом и решает
куда доваить элемент (в левое или правое поддерево). Перепишите, метод
PrintTree для печати полной версии дерева.

```python
import time
import numpy as np

class Tree:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def PrintTree(self, vertex, i=0):
        if vertex is None:
            return None

        print(3 * i * " " + f"[{vertex.data}", end=" ")
        print(f"left: {vertex.left}") #, end=f"{'':>4}")
        self.PrintTree(vertex.left, i=i+1)
        print(3 * i * " " + f"right: {vertex.right}")#, end=f"{'':>4}")
        self.PrintTree(vertex.right, i=i+1)


    def insert(self, data):
        vertex = self
        while True:
            if vertex.data < data:
                if vertex.right == None:
                    vertex.right = Tree(data)
                    break
                vertex = vertex.right

            elif vertex.data > data:
                if vertex.left == None:
                    vertex.left = Tree(data)
                    break
                vertex = vertex.left
            else:
                print("Такой узел уже существует")
                break

```

```python
tree = Tree(5)
tree.insert(6)
tree.insert(56)
tree.insert(2)
tree.insert(3)
tree.insert(5)
tree.insert(0)
tree.PrintTree(tree)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

data = pd.read_csv('https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv')
data
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, -1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
regressor = DecisionTreeClassifier()
regressor.fit(x_train, y_train)
```

```python
tree.plot_tree(regressor)
```

```python
y_pred = regressor.predict(x_test)

predictedData = pd.DataFrame({'True': y_test, 'Predicted': y_pred})
predictedData
```

```python
from sklearn import metrics
print("Mean Squared Error:", metrics.mean_squared_error(y_test, y_pred))
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test, y_pred))
```
