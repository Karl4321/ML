```python
from sklearn import preprocessing
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/akmand/datasets/master/iris.csv")
```

```python
data
```

```python
minMaxScaler=preprocessing.MinMaxScaler(feature_range=(0,1))
scaledData=minMaxScaler.fit_transform(data.drop(columns="species"))
scaledData[:5]
```

```python
ZScaler=preprocessing.StandardScaler()
scaledData=ZScaler.fit_transform(data.drop(columns="species"))
scaledData[:5]
```
