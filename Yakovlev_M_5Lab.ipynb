```python
class TrigonometricFunc:
    def cos(self,x):
        return np.cos(x)
    def sin(self,x):
        return np.sin(x)
    def tg(self,x):
        return np.tan(x)
    def arcsin(self,x):
        return np.arcsin(x)
    def arccos(self,x):
        return np.arccos(x)
    def arctg(self,x):
        return np.arctan(x)
    def GradToRad(self,x):
        return np.radians(x)
```

```python
class Tree:
    def __init__(self,left,right,data):
        self.left=left
        self.right=right
        self.data=data
    def PrintTree(self):
        print(self.data)
        if self.left==None and self.right==None:
            print("---------------------")
        if self.left!=None:
            self.left.PrintTree()
        if self.right!=None:
            self.right.PrintTree()

class IncertToTrees:
    def incert(self,value,originTree):
        if originTree.data<=value:
            return
        if originTree.left==None:
            partTree=originTree.left
            originTree.left=Tree(None,partTree,value)
            return

        if originTree.left.data>=value:
            if originTree.right==None or originTree.right.data<value:
                partTree=originTree.right
                originTree.right=Tree(None,partTree,value)
                return
            if originTree.right.data>value:
                self.incert(value,originTree.right)
                return
        
        if value==originTree.left.data and value==originTree.right.data:
            print("  ")
            return
    
        partTree=originTree.left
        originTree.left=Tree(partTree,None,value)

        
```

```python
a=Tree(Tree(None,None,3),Tree(None,None,4),5)
incert=IncertToTrees()
incert.incert(4,a)
a.PrintTree()
```
